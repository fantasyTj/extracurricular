```python
from glob import glob
import pandas as pd
csv_list = glob('data/*.csv')
df = pd.concat([pd.read_csv(f) for f in csv_list])
df.columns
```

```python
df['summary'] = ''
prompt = '''
近期，全国出现了几起流浪狗伤人事件，激起了网友对于如何对待或处理流浪狗的激烈讨论。
我收集了国内著名问答平台“知乎”相关问题下的回答，现需要请你总结这些问答的主要观点。
问题的回答中可能包含了一些无关的内容，你需要将其去除；回答有可能使用了隐晦的表达方式（如类比），你需要将其转换为直接的表达方式。
下面给出问题和回答：

问题：{question}
回答：{answer}

请你将回答中的主要观点总结出来，总结不超过80字：
'''
max_input_length = 8000 - len(prompt.format(question='问题', answer='回答'))
for index, row in df[:10].iterrows():
    question, answer = row['question_title'], row['content']
    if len(question) + len(answer) > max_input_length:
        answer = answer[:max_input_length - len(question)]
    inner_prompt = prompt.format(question=question, answer=answer)
    df.loc[index, 'summary'] = ' '
```
